```python
import pandas as pd
import energy_utils as etl
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [30, 5]
```

```python
df = pd.read_pickle("./data/shelly bathroom fan 2020-06-13 2021-12-23.pkl")
raw_energy = df.energy.dropna().div(60)
cumul = etl.cumulate(raw_energy)
daily = etl.daily(cumul)
monthly = etl.monthly(daily)
monthly.plot(kind="bar")
```

```python
monthly_json = etl.monthly_json(monthly)
print(monthly_json)
```
