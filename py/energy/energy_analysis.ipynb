```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [30, 5]
```

* Shelly has the energy in Watt Minute => Wh /60
* tuya Zigbee has the energy in Watt Hour => Wh

# Import

```python
df = pd.read_pickle("./data/shelly bathroom fan 2020-06-13 2021-12-23.pkl")
raw_energy = df.energy.dropna().div(60)
raw_energy.plot()
```

# Cumulate Energy through resets

```python
total_energy = []
total_energy_index = []
history = 0
prev = 0
for index, item in raw_energy.iteritems():
    if(item < prev):
        history += prev
    prev = item
    total_energy_index.append(index)
    total_energy.append(item+history)
```

```python
en_df = pd.Series(total_energy, index=total_energy_index)
en_df.plot()
```

## principle for computing the enery of one Day

```python
en_df_days = en_df["2020-06-13"]
en_df_days.plot()
diff_day = en_df_days[-1] - en_df_days[0]
print(f"day energy = {str(round(diff_day, 2))} Wh")
```

## Daily Energy

```python
days_list = []
days_energy_list = []
for idx, day in en_df.groupby(en_df.index.date):
    diff_day = day[-1] - day[0]
    days_list.append(idx)
    days_energy_list.append(diff_day)

days_energy = pd.Series(days_energy_list, index = pd.to_datetime(days_list))
```

```python
days_energy.plot()
```

## Monthly Energy

```python
months_groups = days_energy.groupby(pd.Grouper(freq='1M')).sum()
months_groups.index = months_groups.index.strftime('%B %Y')
months_groups.plot(kind="bar")
```
